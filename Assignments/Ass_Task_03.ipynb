```python
# Allow user to insert hotel name.
# If hotel not found print No hotel found in database
# If found print bar graph of rating count of that hotel.

import matplotlib.pyplot as plt
hotel_name = input("Enter hotel name: ")

if hotel_name not in df['name'].values:
    print("No hotel found in database")
else:
    ratings_count = df[df['name'] == hotel_name]['reviews.rating'].value_counts()
    ratings_count.plot(kind='bar')
    plt.xlabel("Ratings")
    plt.ylabel("Count")
    plt.title(f"Rating Count for {hotel_name}")
    plt.show()
```
