```python
# Task 1
import pandas as pd
import matplotlib.pyplot as plt

# Read the two Excel files into Pandas DataFrames
df1 = pd.read_excel("RESULT1.xlsx")
df2 = pd.read_excel("RESULT2.xlsx")

# Merge the two DataFrames
df_merged = pd.concat([df1, df2])

# Prepare data for bar chart
labels = df_merged['NAME']
values = df_merged['TOTAL']

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, values)

# Add title and labels
plt.title("Student Result Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
```

```python
#task 2a

import requests
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt

#1 covid
url1 = requests.get('https://api.covid19india.org/data.json')
data = url1.json()

dates = []
cases = []


for i in range(0,len(data["cases_time_series"])):
    dates.append(data["cases_time_series"][i]["date"])
    cases.append(int(data["cases_time_series"][i]["dailyconfirmed"]))

#Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(dates, cases)

# Add title and labels
plt.title("Covid-19 Cases in India")
plt.xlabel("Date")
plt.ylabel("Number of Cases")

```

```python
#Task 3


import pandas as pd
import matplotlib.pyplot as plt
import requests
import numpy as np
import json

url = 'https://api.covid19india.org/data.json'
response = requests.get(url)
data = response.json()

state_names = []
cases = []

for i in range(1,len(data['statewise'])):
  state_names.append(data['statewise'][i]["state"])
  cases.append(int(data['statewise'][i]["confirmed"]))

# Create a DataFrame
df = pd.DataFrame({'State': state_names, 'Confirmed Cases': cases})

# Create a horizontal bar chart"
plt.figure(figsize=(10, 6))
plt.barh(df['State'], df['Confirmed Cases'])

# Add title and labels
plt.title('Statewise Confirmed Cases in India')
plt.xlabel('Number of Cases')
plt.ylabel('State')

# Show the plot
plt.show()
```
