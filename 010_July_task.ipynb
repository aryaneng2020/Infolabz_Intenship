```python
import cv2

vid = cv2.VideoCapture(0)
i = 1

while(True):
    ret, frame = vid.read()
    cv2. imshow('frame' , frame)
    
    if cv2.waitKey(1) == ord('q'):
        break
        
    # image store -> machine learning
    # capturing the image and storing it in folder
    elif cv2.waitKey(1) == ord('r'):
        cv2. imwrite('data_{}.jpg'. format(i), frame)
        i=i+1
        continue
vid.release()
cv2. destroyAllWindows ()
```

```python
### Task 2 & 3
import cv2

vid = cv2.VideoCapture(0)
i = 1

while(True):
    ret, frame = vid.read()
    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('Gray Scale Camera', gray_frame)
    
    if cv2.waitKey(1) == ord('q'):
        break
        
    # image store -> machine learning
    # capturing the image and storing it in folder
    elif cv2.waitKey(1) == ord('c'):
        cv2. imwrite('data_gray{}.jpg'. format(i), gray_frame)
        i=i+1
        continue
vid.release()
cv2. destroyAllWindows ()

```

```python
import os
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt

# Define the path to the directory containing the images
dir_path = "imagedata/training/laptop/"

# Iterate over all files in the directory
for img_file in os.listdir(dir_path):
    # Construct the full path to the image file
    img_path = os.path.join(dir_path, img_file)
    
    # Load the image
    img = image.load_img(img_path)
    
    # Display the image
    plt.imshow(img)
    plt.title(img_file)
    plt.axis('off')  # Hide axes for a cleaner look
    plt.show()
```
